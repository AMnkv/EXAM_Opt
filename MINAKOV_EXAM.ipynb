```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
a, b = np.random.randint(-10,10,5), 2
X = np.random.randn(40,5)
y = np.logical_and(X @ a + b > 0, 1).astype(int)
```

```python
# МНК

def MLS_lin_reg(beta,args): #функция которая воспроизводит МНК для нашей функции
    #ее же мы будем оптимизировать методом BFGS
    x,y = args[0],args[1]
    return np.sum((y - np.sum(np.hstack((np.ones([x.shape[0],1]),x)) * beta,axis = 1))**2)

def MLS_lin_reg_gradient(beta,args): #градиент для функции выше
    x,y = args[0],args[1]
    gradient = np.zeros(len(beta))
    gradient[0] = -2*np.sum(y - np.sum(np.hstack((np.ones([x.shape[0],1]),x)) * beta,axis = 1))
    for i in range(x.shape[1]):
        gradient[i+1] = -2*np.sum((y - np.sum(np.hstack((np.ones([x.shape[0],1]),x)) * beta,axis = 1))*x[:,i])

    return gradient
   
        
```

```python
beta = np.array([0,0,0,0,0,0]) #первый элемент в списке это b, так как я добавляю вектор с единицами в начало
res = minimize(MLS_lin_reg, x0 = beta, method='BFGS',
               jac=MLS_lin_reg_gradient, args= [X,y],
               options={'disp': True,'gtol': 1e-6})
```

```python
res.x #наши параметры, первый b, потом вектор а
```

```python
#вектор реальных коэффициентов
beta_real = np.array([b] + a.tolist())
beta_real
```

```python
#разница между реальным и полученным вектором коэффициентов
(res.x - beta_real)
```

```python
#сравним значение MLS для коэффициентов полученных и реальных
print('MLS для настоящих коэффициентов =',MLS_lin_reg(beta_real,[X,y]))

print('MLS для наших коэффициентов =',MLS_lin_reg(res.x ,[X,y]))
```

```python
#теперь все это загоним в функцию
def solve(x,y):
    beta = np.zeros(X.shape[1]+1)
    res = minimize(MLS_lin_reg, x0 = beta, method='BFGS',
               jac=MLS_lin_reg_gradient, args= [X,y],
               
               options={'disp': True,'gtol': 1e-6})
    b = res.x[0]
    a = res.x[1:]
    return a,b
```
